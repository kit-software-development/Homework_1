```python
#http://scikit-learn.org/stable/modules/classes.html#module-sklearn.datasets
from sklearn import datasets

#http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html#sklearn.linear_model.LinearRegression
from sklearn import linear_model
```

```python
# вариант
v = 42

# случайное число в диапозоне 1-20
rnd = random.randint(1,20)
```

### Generate dataset

```python
#http://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_regression.html#sklearn.datasets.make_regression
data, target, coef = datasets.make_regression(n_samples = v * 10, n_features = v, n_informative = round(v/5), n_targets = 1, 
                                              noise = rnd, coef = True, random_state = 2)  
```

### Feature plots

```python
pyplot.figure(figsize(20, 20))

plot_number = 0

for plot_number in range(v):
    pyplot.subplot(10, 5, plot_number + 1)
    pyplot.scatter(list(map(lambda x : x[plot_number], data)), target, color = 'b')
    pyplot.title('feature {}'.format(plot_number))
    pyplot.xlabel('feature {}'.format(plot_number))
    pyplot.ylabel(target)
    plot_number += 1
```

### Linear regression model learning

```python
linear_regressor = linear_model.LinearRegression()
linear_regressor.fit(data, target)
predictions = linear_regressor.predict(data)
```

### Print learned and true regression models

```python
print ("Learned regression model")
string = "y = {:.3f}".format(linear_regressor.intercept_)
for i in range(len(linear_regressor.coef_)):
    string += " + {:.3f}*x{}".format(linear_regressor.coef_[i], i+1)

print(string)
```

```python
print ("True regression model")
string = "y = "

for i in range(len(coef)):
    if (round(coef[i],3)==0):
        continue
    string += "+ {:.3f}*x{} ".format(coef[i], i+1)
    
print(string)
```
